```python
import gensim
import pandas as pd
```

```python
df=pd.read_json("/content/Cell_Phones_and_Accessories_5.json",lines=True)
df.head(5)
```

```python
df.reviewText[0]
```

```python
df.shape
```

```python
# simple preprocess by gensim module utils and creating new column
df["processed_txt"]=df.reviewText.apply(gensim.utils.simple_preprocess)
```

```python
# df.processed_txt[0]
```

```python
# creating model
model=gensim.models.Word2Vec(window=10,min_count=2,workers=4)
# creating vocabulary the model
model.build_vocab(df.processed_txt,progress_per=1000)
```

```python
# training the model
model.train(df.processed_txt,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.save("/content/word2vec.model")
```
