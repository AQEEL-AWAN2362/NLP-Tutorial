```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
doc=nlp("Apple is looking at buying U.K. startup for $1 billion") # token.tag_ explain the tense
for token in doc:
    print(token, " | ", token.pos_, " | ",spacy.explain(token.pos_), " | ",token.tag_, " | ",spacy.explain(token.tag_))
```
