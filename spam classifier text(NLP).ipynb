```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
from nltk import sent_tokenize,word_tokenize
import gensim
from gensim.utils import simple_preprocess # simple preproces (lower case , tokinize into each words)
```

```python
# cleaning data
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z0-9]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review if not word in list(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# tokenize into sentences part
words=[]
for sent in messages["message"]:
    sent_token=sent_tokenize(sent)
    for senten in sent_token:
        words.append(simple_preprocess(senten))
```

```python
words
```

```python
# tokenize into words
words=[]
for word in messages["message"]:
    word_token=word_tokenize(word)
    for word in word_token:
        words.append(simple_preprocess(word))
```

```python
words
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
X = cv.fit_transform(corpus).toarray()
```
