```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))  # Download once, use everywhere
```

```python
# stemming
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    stop_words = set(stopwords.words("english"))
    if word not in stop_words:
      print(stemmer.stem(word))
```

```python
# lemmatization
for i in corpus:
  words=nltk.word_tokenize(i)
  for word in words:
    stop_words = set(stopwords.words("english"))
    if word not in stop_words:
      print(lemmatizer.lemmatize(word))
```

```python
# all in one
import re
corpus=[]
for i in range(len(sentences)):
  review=re.sub("[^a-zA-Z]"," ",sentences[i])
  review=review.lower()
  review=review.split()# converting into tokens
  review=[lemmatizer.lemmatize(word) for word in review if not word in stop_words]
  review=" ".join(review)# converting back the list into a cleaned string.
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary=True)
```

```python
x=cv.fit_transform(corpus)
cv.vocabulary_
```
