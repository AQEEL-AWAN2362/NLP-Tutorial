```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus=[
    "Thor eating pizza, Loki is eating pizza, Ironman ate pizza already",
    "Apple is announcing new iphone tomorrow",
    "Tesla is announcing new model-3 tomorrow",
    "Google is announcing new pixel-6 tomorrow",
    "Microsoft is announcing new surface tomorrow",
    "Amazon is announcing new eco-dot tomorrow",
    "I am eating biryani and you are eating grapes"
]
```

```python
tf=TfidfVectorizer()
transformed_output=tf.fit_transform(corpus)
tf.vocabulary_
```

```python
dir(tf)
```

```python
# getting scores of each words
all_feature_names=tf.get_feature_names_out()
for word in all_feature_names:
    indx=tf.vocabulary_.get(word)
    print(f"{word} {tf.idf_[indx]}")
```

```python
import pandas as pd
df=pd.read_csv("/content/Ecommerce_data.csv")
df.head()
```

```python
# checking the balancing data
df.label.value_counts()
```

```python
# encode the label column
df["label_num"]=df.label.map({"Household":0,
              "Electronics":1,
              "Books":2,
              "Clothing & Accessories":3
})

df.head()
```

```python
# train test splitt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    df.Text,
    df.label_num,
    test_size=0.2,
    stratify=df.label_num)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
x_train.shape, x_test.shape
```

```python
# checking the equal distribution of samples in y
y_train.value_counts(), y_test.value_counts()
```

```python
# randomforest classifier
clf=Pipeline([
    ("vectorizer_tfidf",TfidfVectorizer()),
    ("random_forest",RandomForestClassifier())
])

clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
# Naive bayes classifier
clf=Pipeline([
    ("vectorizer_tfidf",TfidfVectorizer()),
    ("mnnb",MultinomialNB())
])

clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
# knc
clf=Pipeline([
    ("vectorizer_tfidf",TfidfVectorizer()),
    ("knc",KNeighborsClassifier()),
])

clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
# using with preprocess data
import spacy
nlp=spacy.load("en_core_web_sm")
def preprocess(text):
  doc=nlp(text)
  filtered_tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
  print(filtered_tokens)
  return " ".join(filtered_tokens)
```

```python
df["preprocessed txt"]=df['Text'].apply(preprocess)
```

```python
# after preprocess train test splitt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    df['preprocessed txt'],
    df.label_num,
    test_size=0.2,
    stratify=df.label_num)
```

```python
# knc
clf=Pipeline([
    ("vectorizer_tfidf",TfidfVectorizer()),
    ("knc",KNeighborsClassifier()),
])

clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
# randomforest classifier
clf=Pipeline([
    ("vectorizer_tfidf",TfidfVectorizer()),
    ("random_forest",RandomForestClassifier())
])

clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(classification_report(y_test,y_pred))
```
