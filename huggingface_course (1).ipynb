**Preprocessing on data**

```python
# checking short reviews
short_reviews = data["train"].filter(lambda x: len(x["text"]) < 100)
short_reviews
```

```python
# add word count column of text
def add_word_count(text: str)  :
  text["add_word_count"] = len(text["text"].split())
  return text
data1 = short_reviews.map(add_word_count)
print(data1)
data1.column_names
data1[0]



```

```python
# splitting the train dataset into train and test
train_dataset = data["train"].train_test_split(test_size=0.2)
train_dataset
```
